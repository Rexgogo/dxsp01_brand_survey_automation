```python
from pathlib import Path
from rich.tree import Tree
from rich.console import Console

def build_tree(path: Path, tree: Tree):
    """éè¿´æ§‹å»ºæ¨¹ç‹€çµæ§‹"""
    for item in sorted(path.iterdir()):
        if item.name.startswith("."):  # å¿½ç•¥éš±è—æª”èˆ‡ .git ç­‰
            continue
        if item.is_dir():
            branch = tree.add(f"ğŸ“ {item.name}")
            build_tree(item, branch)
        else:
            tree.add(f"ğŸ“„ {item.name}")

# âœ¨ è«‹å°‡é€™è£¡æ›æˆä½ å¯¦éš›çš„å°ˆæ¡ˆè³‡æ–™å¤¾è·¯å¾‘
project_root = Path("D:/Rex/Desktop/DS_workspace/dxsp01_brand_survey_automation")

# æª¢æŸ¥è·¯å¾‘æ˜¯å¦å­˜åœ¨
if not project_root.exists():
    print(f"[red]âŒ æ‰¾ä¸åˆ°è·¯å¾‘ï¼š{project_root}ï¼Œè«‹ç¢ºèªæ˜¯å¦æ­£ç¢ºã€‚[/red]")
else:
    # åˆå§‹åŒ– Tree å’Œ Console
    root_tree = Tree(f"ğŸ“ {project_root.name}")
    build_tree(project_root, root_tree)

    console = Console()
    console.print(root_tree)

    # âœ¨ å°‡çµæœè¼¸å‡ºç‚º Markdown æ ¼å¼çš„æª”æ¡ˆ
    docs_path = project_root / "docs"
    docs_path.mkdir(parents=True, exist_ok=True)  # ç¢ºä¿ docs è³‡æ–™å¤¾å­˜åœ¨

    file_output = docs_path / "file_system_diagram.md"
    
    # ä½¿ç”¨ Console.capture() æ“·å–æ¨¹ç‹€åœ–è¼¸å‡º
    with console.capture() as capture:
        console.print(root_tree)
    file_output.write_text(capture.get(), encoding='utf-8')

    print(f"[green]âœ… å·²å°‡æª”æ¡ˆçµæ§‹å¯«å…¥ï¼š{file_output}[/green]")
```
